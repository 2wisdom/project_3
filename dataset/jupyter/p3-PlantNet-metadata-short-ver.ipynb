```python
# JSON파일을 LIST형식으로 변환
parse_list = [] 
header = [
    "id", 
    "species_id",
    "obs_id",
    "organ",
    "is_shareable",
    "v1_id",
    "author",
    "license",
    "split",
]
for key, value in json_data.items():
    item = []
    item.append(key)

    for k, v in value.items():
        item.append(v)

    parse_list.append(item)
parse_list
```

```python
df_list = pd.DataFrame(parse_list, columns=[header])
df_list.head()
```
